# EDA

```python
plt.figure(figsize=(15,6))
plt.bar(x=list(df['Company Name'].value_counts().index), height=list(df['Company Name'].value_counts()),color='r')
plt.xticks(rotation=90)
plt.xlabel('Company',fontsize=14)
plt.ylabel('Number of Missions',fontsize=14)
plt.title('Total Number Of Missions Conducted By Each Company',fontsize=18)
plt.show()
```

```python
m=[]
for x in df['Country'].unique():
    n=[]
    df_c=df[df['Country']== x]
    for y in df['Year'].unique():
        df_y=df_c[df_c['Year']==y]
        n.append(df_y.count()['Year'])
    m.append(n)
pivot=pd.DataFrame(data=m, columns=df['Year'].unique(), index=df['Country'].unique())
plt.figure(figsize=(20,10))
sns.heatmap(pivot,cmap='coolwarm',linecolor='white', linewidth= 1)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Country', fontsize=18)
plt.title('Number Of Missions In Each Country In Each Year', fontsize=25)
plt.show()
```

```python
df_loc=df.groupby(by='Location').count().sort_values(by='Company Name', ascending=False).head(15)
plt.figure(figsize=(4,4))
plt.bar(x=list(df_loc.index), height=df_loc['Company Name'],color='turquoise')
plt.xlabel('Location', fontsize=14)
plt.ylabel('Number of Missions', fontsize=14)
plt.xticks(rotation=90)
plt.title('Top 15 Locations With The Highest Number Of Missions', fontsize=18)
plt.show()
```

```python
df_country_10=df.groupby('Country').count().sort_values(by='Company Name',ascending=False).head(5)
plt.figure(figsize=(5,10))
plt.pie(x=df_country_10['Company Name'],labels=list(df_country_10.index), pctdistance=0.8, labeldistance=1.05,autopct='%1.0f%%',explode=(0.1,0,0,0,0),shadow=True,startangle=0,colors=['limegreen', 'deeppink', 'deepskyblue', 'orange','teal'],textprops={'fontsize':14})
plt.title('Top 5 Countries Involved In Space Missions', fontsize=15)
plt.show()
```
